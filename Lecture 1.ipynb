```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Practive").getOrCreate()
df = spark.table("default.data_csv")
```

```
df.show()
```

```
from pyspark.sql.functions import max
max_age = df.agg(max(df['Age'])).collect()[0][0]

# Print the maximum age
print("Maximum Age:", max_age)
```

```
filtered_df = df.filter(df['Age'] >= 30)
filtered_df.show()
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Practive").getOrCreate()
prescriptions = spark.table("default.prescriptions_csv")
```

```
prescriptions.show(5)
```

```
from pyspark.sql.functions import countDistinct
num_studies = prescriptions.agg(countDistinct("BNF_CODE").alias("Number_of_Studies")).collect()[0]["Number_of_Studies"]
 
# Print the result
print("The number of Medicines in the dataset is:", num_studies)
 
```

```
from pyspark.sql.functions import count, desc
type_counts = prescriptions.groupBy("PRACTICE_CODE").agg(count("*").alias("Frequency"))
# Sort by frequency in descending order
type_counts = type_counts.sort(desc("Frequency"))
# Show the results
type_counts.show()
```

```
from pyspark.sql.functions import sum
sum_cost_df = prescriptions.groupBy('BNF_CODE').agg(sum('ACTUAL_COST').alias('Total_Cost'))

# Show the result
sum_cost_df.show()
```

```
from pyspark.sql.functions import sum
sum_cost_df = prescriptions.groupBy('BNF_CODE').agg(sum('ACTUAL_COST').alias('Total_Cost')).orderBy('Total_Cost', ascending=False)

# Show the result
sum_cost_df.show()
```
